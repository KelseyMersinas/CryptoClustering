```python
# import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import holoviews as hv
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# load in the csv
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# display the data 
df_market_data.head(10)
```

```python
# generate summary statistics
df_market_data.describe()
```

```python
# plot df_market_data to show what is in the DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# Use `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y']])
```

```python
# Copy the crypto names from the original data
coin_id = df_market_data.index

# Create a DataFrame with the scaled data and set the coin ids as the index
market_data_transformed = pd.DataFrame(market_data_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'], index=coin_id)

# Display sample data
market_data_transformed.head(10)
```

### Find the Best Value for K Using the Original Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(market_data_transformed)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

##### Question: What is the best value for k?
Answer: From this data and resulting elbow graph the best answer is 4. 


### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(market_data_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4_predictions = model.predict(market_data_transformed)

# Print the resulting array of cluster values.
print(k_4_predictions)
```

```python
# Create a copy of the DataFrame
market_data_predictions = market_data_transformed.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_data_predictions['price clusters'] = k_4_predictions

# Display sample data
market_data_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
market_data_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="price clusters"
)
```

### Optimize Clusters with Principal Component Analysis

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca = pca.fit_transform(market_data_transformed)

# View the first five rows of the DataFrame. 
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

##### Question: What is the total explained variance of the three principal components?
Answer: The variance of the three components are 89.7% This shows that theses three components are a good choice to retain the majority of our data. 

```python
# Create a new DataFrame with the PCA data.
market_data_pca_df = pd.DataFrame(
    market_data_pca,
    columns=['PCA1', 'PCA2', 'PCA3']
)

# Copy the crypto names from the original data
market_data_pca_df.index = coin_id

# Display sample data
market_data_pca_df.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(market_data_pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
df_PCA_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_PCA_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

##### Question: What is the best value for k when using the PCA data?
Answer: With this elbow graph using the PCA data, the best answer is 4. when k=4 there is the bigesty change in inertia.

##### Question: Does it differ from the best k value found using the original data?
Answer: No, this is the same number of clusters as previously found. While PCA does reduce the data, the core of the information should be staying the same. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
model.fit(market_data_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_4 = model.predict(market_data_pca_df)
# Print the resulting array of cluster values.
print(k_4)
```

```python
# Create a copy of the DataFrame with the PCA data
market_data_pca_predictions = market_data_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
market_data_pca_predictions['crypto clusters'] = k_4

# Display sample data
market_data_pca_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
market_data_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="crypto clusters"
)
```

### Visualize and Compare the Results

```python
# composite plots to show and contrast both elbow curves
# combine the plots
original_elbow = df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title='Elbow Curve', 
    label='Original Elbow',
    xticks=k
)

pca_elbow = df_PCA_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title='Elbow Curve', 
    label='PCA Elbow',
    xticks=k
)
combined_elbow = original_elbow * pca_elbow

# label
combined_elbow = combined_elbow.opts(
    title='Composite Elbow Curves',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    legend_position='right'
)

# show
combined_elbow
```

```python
# Composite plot to contrast the clusters
# combine the plots
scatter_original = market_data_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="price clusters",
    title="Clusters in Original Feature Space",
    color='blue',
    legend_label='Original Clusters'
)

scatter_pca = market_data_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="crypto clusters",
    title="Clusters in PCA-transformed Space",
    color='green',
    legend_label='PCA Clusters'
)

# overlay
combined_scatter = scatter_original * scatter_pca

# label
combined_scatter = combined_scatter.opts(
    title="Composite Scatter Plots",
    xlabel='price_change_percentage_24h',  
    ylabel='price_change_percentage_7d', 
    legend_position='right'
)

# show
combined_scatter
```
